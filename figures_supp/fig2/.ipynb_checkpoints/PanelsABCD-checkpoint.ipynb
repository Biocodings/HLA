## Panel B

```python
def get_similarity(caller1, caller2):
    merged = pd.merge(caller1, caller2, on='Sample')
    all_samples, all_similarities = [], []
    for row in merged.iterrows():
        if (row[1]['A1_x'] == '-')|(row[1]['A1_y'] == '-'):
            continue
        similar = 0
        if ((row[1]['A1_x'] == row[1]['A1_y'])&(row[1]['A2_x'] == row[1]['A2_y']))|((row[1]['A1_x'] == row[1]['A2_y'])&(row[1]['A2_x'] == row[1]['A1_y'])):
            similar += 2
        elif ((row[1]['A1_x'] == row[1]['A1_y'])|(row[1]['A2_x'] == row[1]['A2_y']))|((row[1]['A1_x'] == row[1]['A2_y'])|(row[1]['A2_x'] == row[1]['A1_y'])):
            similar += 1
        if ((row[1]['B1_x'] == row[1]['B1_y'])&(row[1]['B2_x'] == row[1]['B2_y']))|((row[1]['B1_x'] == row[1]['B2_y'])&(row[1]['B2_x'] == row[1]['B1_y'])):
            similar += 2
        elif ((row[1]['B1_x'] == row[1]['B1_y'])|(row[1]['B2_x'] == row[1]['B2_y']))|((row[1]['B1_x'] == row[1]['B2_y'])|(row[1]['B2_x'] == row[1]['B1_y'])):
            similar += 1
        if ((row[1]['C1_x'] == row[1]['C1_y'])&(row[1]['C2_x'] == row[1]['C2_y']))|((row[1]['C1_x'] == row[1]['C2_y'])&(row[1]['C2_x'] == row[1]['C1_y'])):
            similar += 2
        elif ((row[1]['C1_x'] == row[1]['C1_y'])|(row[1]['C2_x'] == row[1]['C2_y']))|((row[1]['C1_x'] == row[1]['C2_y'])|(row[1]['C2_x'] == row[1]['C1_y'])):
            similar += 1
        all_similarities.append(similar)
        all_samples.append(row[1]['Sample'])
    df = pd.DataFrame({"Samples": all_samples, "Similarity": all_similarities})
    return df
```

```python
optitype_polysolver = get_similarity(optitype, polysolver)
```

```python
sns.countplot(x='Similarity', data=optitype_polysolver, color='white')
plt.title('Optitype and Polysolver')
plt.ylabel('Number of patients')
plt.xlabel('Number of identically typed HLA alleles')
```

## Panel C <br>
Data required: genotype information about patients typed with snp2hla

```python
df = pd.read_csv('/cellar/users/ramarty/Projects/hla_new/data/typing_the_hla/plink.mds.usedforfig1.txt', delim_whitespace=True)
```

```python
def get_tcga(x):
    if x[:4] == 'TCGA':
        return True
    else:
        return False
df['TCGA'] = df['FID'].apply(get_tcga)
```

```python
maps = pd.read_csv('/cellar/users/ramarty/Projects/hla_new/data/typing_the_hla/hapmap', sep='\t')
```

```python
df2 = pd.merge(maps, df, on='IID', how='inner')
df2['population'] = df2['population'].fillna('TCGA')
```

```python
tcga = df[df.TCGA]
tcga['population'] = 'TCGA'
```

```python
df3 = pd.concat([df2[['population', 'C1', 'C2']], tcga[['population', 'C1', 'C2']]])
```

```python
def clear_pops(x):
    if x == 'TCGA':
        return 'TCGA'
    elif x == 'JPT' or x == 'CHB' or x == 'CHD':
        return 'Asian'
    elif x == 'CEU':
        return 'European'
    elif x == 'ASW' or x == 'YRI' or x == 'LWK' or x == 'MKK':
        return 'African'
    elif x == 'GIH' or x == 'MXL':
        return 'Indian/Mexican'
df3['pop'] = df3['population'].apply(clear_pops)
```

```python
current_palette = sns.color_palette()
sns.lmplot(x='C1', y='C2', hue='pop', data=df3, fit_reg=False, size=10,
          hue_order=['African', 'Asian', 'Indian/Mexican', 'TCGA', 'European'])
```

## Panel D

```python
inconsistent_samples = list(optitype_polysolver[optitype_polysolver.Similarity < 5].Samples)
```

```python
samples_from_optitype_and_polysolver = list(set(list(optitype[~optitype.Sample.isin(inconsistent_samples)].Sample) + list(polysolver[~polysolver.Sample.isin(inconsistent_samples)].Sample)))
```

```python
to_add_from_snp2hla = [x for x in snp2hla.Sample if x not in samples_from_optitype_and_polysolver]
to_add_from_polysolver = [x for x in list(polysolver[~polysolver.Sample.isin(inconsistent_samples)].Sample) if x not in list(optitype[~optitype.Sample.isin(inconsistent_samples)].Sample)]
to_add_from_optitype = [x for x in optitype[~optitype.Sample.isin(inconsistent_samples)].Sample if x not in list(polysolver[~polysolver.Sample.isin(inconsistent_samples)].Sample)]
to_add_from_consensus = [x for x in patients_with_mutations if (x not in to_add_from_optitype)&(x not in to_add_from_polysolver)&(x not in to_add_from_snp2hla)]
```

```python
included_count_df = pd.DataFrame({'Tool': ['Opti_Poly_consensus', 'Polysolver', 'Optitype', 'Snp2HLA'],
                                  'Patients_typed': [len(to_add_from_consensus), len(to_add_from_polysolver), 
                                                     len(to_add_from_optitype), len(to_add_from_snp2hla)]}) 
sns.barplot(x='Tool', y='Patients_typed', data=included_count_df, color='white')
```
